## Line plots

Let's take a look at creating a simple *line plot*.

* **plt.plot()**: plot ``y`` versus ``x`` as lines and/or markers. 



```python
x = np.linspace(0, 10, 100)

fig = plt.figure()
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))

plt.show()
```

### Line colors and styles

The ``plt.plot()`` function takes the optinal parameters that defines basic formatting like *color*, *marker* and *linestyle*. 

* ``color``: a color can be specified in a variety of ways.
  * If no color is specified, Matplotlib will automatically cycle through a set of default colors for multiple lines. 
  * Here are a list of named colors: https://matplotlib.org/devdocs/gallery/color/named_colors.html



```python
fig = plt.figure()

plt.plot(x, np.sin(x - 0), color='blue')        # specify color by name
plt.plot(x, np.sin(x - 1), color='g')           # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75')        # Grayscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44')     # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple, values 0 to 1
plt.plot(x, np.sin(x - 5), color='blueviolet'); 

plt.grid()
plt.show()
```

* ``linestyle`` or ``ls``: 
  * Linestyles can be defined using the strings, or a code. 
  * `'-'` or `'solid'`: soild line
  * `'--'` or `'dashed'`: dashed line
  * `'-.'` or `'dashdot'`: dash-dotted line
  * `':'` or `'dotted'`: dotted line
  * https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html?highlight=linestyle


```python
fig = plt.figure()

plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');

plt.plot(x, x + 4, linestyle='-')  # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':');  # dotted

plt.grid()
plt.show()
```

* ``linewidth`` or ``lw``: 
  * line width in points. 

```python
fig = plt.figure()

plt.plot(x, x + 0, linestyle='-', linewidth=0.5)  
plt.plot(x, x + 2, linestyle='-', linewidth=1) 
plt.plot(x, x + 4, linestyle='-', linewidth=2) 
plt.plot(x, x + 6, linestyle='-', linewidth=3.5)  

plt.grid()
plt.show()
```

* ``marker``: 
  * Marker functionality can be used for plot, scatter, and errorbar. 
  * https://matplotlib.org/stable/api/markers_api.html#module-matplotlib.markers

* ``markersize`` or ``ms``:
  * marker size in points.

```python
x = np.linspace(0,10,11)

fig = plt.figure()

plt.plot(x, x + 0, linestyle='-', marker='o', ms=3)  
plt.plot(x, x + 2, linestyle='-', marker='>', ms=6) 
plt.plot(x, x + 4, linestyle='-', marker='P', ms=9) 
plt.plot(x, x + 6, linestyle='-', marker='*', ms=12);  

plt.grid()
plt.show()
```

```python
x = np.linspace(0, 10, 10)
plt.plot(x, x + 0, 'go--', linewidth=2, markersize=12)
plt.plot(x, x + 2, 'bs-.', linewidth=0.5, markersize=5)

plt.grid()
plt.show()
```

### Axes Limits, Labeling

* ``plt.xlim()``, ``plt.ylim()``: adjust axis limits
  * ``plt.xlim([low, high])``
* ``plt.axis()``: adjust limits of all axis at the same time
  


```python
x = np.linspace(0, 10, 100)
fig = plt.figure()
plt.plot(x, np.sin(x))

plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5)

plt.grid()
plt.show()
```

```python
x = np.linspace(0, 10, 100)
fig = plt.figure()
plt.plot(x, np.sin(x))

plt.axis([-1, 11, -1.5, 1.5])

plt.grid()
plt.show()
```

Labeling of plots includes titles, axis labels, and simple legends. The position, size, and style of these labels can be adjusted using optional arguments.

* ``plt.title()``
* ``plt.xlabel()``, ``plt.ylabel()``
* ``plt.legend()``
  * ``loc``: 'best', 'upper/lower right/left/center', 'left/right/center'

```python
fig = plt.figure()

plt.plot(x, np.sin(x))

plt.title("A Sine Curve")
plt.xlabel("x")
plt.ylabel("sin(x)");

plt.grid()
plt.show()
```

```python
fig = plt.figure()

plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')

#plt.legend();
plt.legend(loc='lower center')

plt.grid()
plt.show()
```

## Creating Multiple Plots on Same Canvas

* ``plt.subplot(grid_position)``: creates a single subplot axes at a specified grid position



```python
x = np.linspace(0,5,11)
fig = plt.figure()

plt.subplot(1,2,1)
plt.plot(x, x**2, 'r--') 
plt.xlabel('(a)')
plt.grid()

plt.subplot(1,2,2)
plt.plot(x, np.sqrt(x), 'g*-');
plt.xlabel('(b)')

plt.grid()
plt.show()
```

## Object Oriented Method

The more formal Object Oriented method is to create figure objects and then just call methods or attributes off of the object. This approach is nicer when dealing with a canvas that has multiple plots on it. 

* ``plt.subplots(array_size)``: it generates a figure object (``fig``) that holds a specific array of subplots (or ``axes`` objects)


* ``ax.plot()``
* ``ax.set_xlabel()``, ``ax.set_ylabel()``, ``ax.set_title()``
* ``ax.grid()``

```python
# Similar to plt.figure(), but grap fig and axes
fig, axes = plt.subplots()

axes.plot(x, x**2, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y = x^2')
axes.set_title('title')
axes.grid()
```

```python
x = np.linspace(0,10,100)

fig, axes = plt.subplots(2)

axes[0].plot(x, np.sin(x))
axes[1].plot(x, np.cos(x));

for ax in axes:
  ax.set_xlabel('x')
  ax.set_ylabel('y')
  ax.set_title('Title')
  ax.grid()
```

```python
x = np.linspace(0, 5, 11)

fig, ax = plt.subplots()

ax.plot(x, x+1, color="red", linewidth=0.25, label='1')
ax.plot(x, x+2, color="red", linewidth=0.50, label='2')
ax.plot(x, x+3, color="red", linewidth=1.00, label='3')
ax.plot(x, x+4, color="red", linewidth=2.00, label='4')

ax.plot(x, x+5, color="green", lw=3, linestyle='-', label='5')
ax.plot(x, x+6, color="green", lw=3, ls='-.', label='6')
ax.plot(x, x+7, color="green", lw=3, ls=':', label='7')

ax.plot(x, x+ 9, color="blue", lw=3, ls='-', marker='+', label='9')
ax.plot(x, x+10, color="blue", lw=3, ls='--', marker='o', label='10')
ax.plot(x, x+11, color="blue", lw=3, ls='-', marker='s', label='11')
ax.plot(x, x+12, color="blue", lw=3, ls='--', marker='1', label='12')

ax.plot(x, x+13, color="purple", lw=1, ls='-', marker='o', markersize=2, label='13')
ax.plot(x, x+14, color="purple", lw=1, ls='-', marker='o', markersize=4, label='14')
ax.plot(x, x+15, color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red", label='15')
ax.plot(x, x+16, color="purple", lw=1, ls='-', marker='s', markersize=8, 
        markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green", label='16');

ax.set_title('Title')
ax.legend(loc='best', ncol=5)

fig.show()
```

```python
# Creates blank canvas
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # main axes
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # inset axes

# Larger Figure Axes 1
axes1.plot(x, x**2, 'b')
axes1.set_xlabel('X_label_axes1')
axes1.set_ylabel('Y_label_axes1')
axes1.set_title('Axes 1 Title')

# Insert Figure Axes 2
axes2.plot(x**2, x, 'r')
axes2.set_xlabel('X_label_axes2')
axes2.set_ylabel('Y_label_axes2')
axes2.set_title('Axes 2 Title');
```

## Scatter Plots

Another commonly used plot type is the simple *scatter* plot. Instead of points being joined by line segments, in scatter plot, the points are represented individually with a dot, circle, or other shape. 

``plt.scatter()`` is used very similar to the ``plt.plot`` but to create scatter plot.  The properties of each inidividual point (e.g., size, face color, edge colkor, etc) can be individually controlled. 

* ``c`` : color
* ``s``: marker size in points
* ``marker``: marker style
* ``cmap``: colormap
* ``alpha``: between 0 (transparent) and 1 (opaque)
* ``edgecolors``: the edge color of the marker



```python
x = np.linspace(0,10,30)
y = np.sin(x)

fig = plt.figure()

plt.scatter(x,y, marker='o')

plt.grid()
plt.show()
```

```python
x = np.random.randn(100)
y = np.random.randn(100)

colors = np.random.rand(100)
sizes = 1000 * np.random.rand(100)

fig = plt.figure()
plt.scatter(x, y, c=colors, s=sizes, alpha=0.3,
            cmap='viridis')
plt.colorbar()
plt.grid()
plt.show()
```

```python
rng = np.random.RandomState(0)

for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(5), rng.rand(5), marker,
             label="marker='{0}'".format(marker))

plt.legend(numpoints=1)
plt.xlim(0, 1.8);
```

```python
X = np.random.normal(0, 1, 100)
Y = np.random.normal(0, 1, 100)
C = np.random.randint(0, 5, 100)

cmap_lst = [plt.cm.viridis, plt.cm.rainbow, plt.cm.Accent, plt.cm.Blues]

f, axes = plt.subplots(1, 4, sharex=True, sharey=True, figsize=(16,4))

for i in range(0, 4):
    axes[i].scatter(X, Y, c=C, cmap=cmap_lst[i])
    axes[i].set_title("cmap: {}".format(cmap_lst[i].name))
plt.show()
```

## Quiz

* Create a figure and axes objects called ``fig``, ``axes`` using ``plt.subplots()``. The subplot has figsize ``(12,4)`` and includes a $1 \times 2$ array of subplots.

* Plot $(x,y)$ on the first subplot and $(x,z)$ on the second subplot. Set line properties to match the plot below.

* Set the labels and titles to match the plot below.

```python
x = np.arange(0, 100, 11)
y = x*2
z = x**2
```

```python
# create a fig, axes objects
fig, axes = plt.subplots(1,2, figsize=(12,4))

# plot (x,y) on the first subplot and (x,z) on the second subplot
axes[0].plot(x,y, 'r-', marker='o')
axes[1].plot(x,z, 'g:', marker='<')

# set labels and title to match the plot below
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].set_title('(a)')

axes[1].set_xlabel('x')
axes[1].set_ylabel('z')
axes[1].set_title('(b)')

axes[0].grid()
axes[1].grid()
fig.show()
```

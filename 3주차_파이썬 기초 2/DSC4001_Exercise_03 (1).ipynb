### Functions

Define a function ``plus(a,b)`` that accepts two numbers ``a`` and ``b`` and return the sum.

```
# define a function plus()
def plus(a,b):
  return a+b
```

```
result = plus(3,5)
print('result of plus(3,5) is %d' %(result))

result = plus(a=3, b=5)
print('result of plus(a=3, b=5) is %d'%(result))
```

```
# define a function plus() with default arguments
def plus(a, b=5):
  return a+b
```

```
result = plus(3,4)
print('result of plus(3,4) is %d' %(result))

result = plus(3)
print('result of plus(3) is %d' %(result))
```

Define a function ``plus(*args)`` that can accept variable number of arguments, and return the sum of all arguments.

```
# variable length of arguments
def plus(*args):
  result = 0

  #print(type(args))
  #print(len(args))
  for i in args:
    result += i

  return result
```

```
plus(1,3,5)
```

Define a function ``plus(**kwargs)``.

```
def plus(**kwargs):
  result = 0

  #print(type(kwargs))
  #print(len(kwargs))
  for key in kwargs.keys():
    result += kwargs[key]
    #print('key: %s, value: %s' %(key, kwargs[key]))

  return result
```

```
plus(a=1, b=3, c=5)
```

Define a function ``plus(a,b)`` that returns multiple values (a+b, a, b).

```
def plus(a,b):
  sum = a + b
  return (sum, a, b)
```

```
result = plus(3,5)
print('result of plus(3,5) is {}'.format(result))
print(result[0], result[1], result[2])

result, aa, bb = plus(3,5)
print('result of plus(3,5) is %d %d %d' %(result, aa, bb))
```

**Handling real data set .csv file**

CSV (Comma Seperated Values) is a simple file format used to store tabular data, such as a spreadsheet or database. CSV file stores tabular data (numbers and text) in plain text. Each line of the file is a data record. Each record consists of one or more fileds, separated by commas. 

```
import os
import tarfile
import urllib.request

DOWNLOAD_ROOT = "https://raw.githubusercontent.com/ageron/handson-ml2/master/"
HOUSING_PATH = os.path.join("datasets", "housing")
HOUSING_URL = DOWNLOAD_ROOT + "datasets/housing/housing.tgz"

def fetch_housing_data(housing_url=HOUSING_URL, housing_path=HOUSING_PATH):
    if not os.path.isdir(housing_path):
        os.makedirs(housing_path)
    tgz_path = os.path.join(housing_path, "housing.tgz")
    urllib.request.urlretrieve(housing_url, tgz_path)
    housing_tgz = tarfile.open(tgz_path)
    housing_tgz.extractall(path=housing_path)
    housing_tgz.close()
```

Calculate the mean of the attribute 'housing_median_age': 1) find the index of the attribute from the first line of the file. 2) from the second line of the file, compute the sum of the 'housing_median_age' value of each line. 3) compute the mean value.

```
with open(HOUSING_PATH+'/housing.csv', 'r') as f:
  lines =f.readlines()
  sum = 0
  keyword = 'housing_median_age'

  line_ref = lines[0].strip().split(',')  
  k_idx = line_ref.index(keyword)
  print(line_ref)
  print('index of the keyword %s is %d' %(keyword, k_idx))

  for lidx in range(1,len(lines)):
    line = lines[lidx]
    line_list = line.strip().split(',')
    
    sum += float(line_list[k_idx])

  print('mean of %s is %f' %(keyword, sum/(len(lines)-1)))
```

Read CSV file using 'csv' module.

```
import csv

lines = []

with open(HOUSING_PATH+'/housing.csv','r') as f:
  reader = csv.reader(f)
  keyword = 'housing_median_age'  
  sum = 0

  for lidx, line in enumerate(reader):
    lines.append(line)

    if lidx == 0:
      k_idx = line.index(keyword)
      print(line)
      print('index of the keyword %s is %d' %(keyword, k_idx))
    else:
      sum += float(line[k_idx])


  print('mean of %s is %f' %(keyword, sum/lidx))
```
